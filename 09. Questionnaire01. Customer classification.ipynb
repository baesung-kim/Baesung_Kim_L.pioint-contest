```python
import numpy as np
import pandas as pd
Back = pd.read_csv('혜택구하기.csv', engine='python', encoding='utf-8-sig')
Back = Back.drop('Unnamed: 0',axis =1)
```

```python
df = Back[['구매경과시간','Back_nums','검색어유사도']]
```

```python
A_list=[]
B_list=[]
C_list=[]


for i in range(df.shape[0]):
    A_list.append(list(df.iloc[i]).count('A'))
    B_list.append(list(df.iloc[i]).count('B'))
    C_list.append(list(df.iloc[i]).count('C'))
```

```python
len(A_list)
```

```python
df['신중도_A'] = A_list
df['신중도_B'] = B_list
df['신중도_C'] = C_list

df['신중도_B'] = df['신중도_B'] * (1.33)
```

```python
df['max_number'] = df[['신중도_A','신중도_B','신중도_C']].max(axis=1)
df
```

```python
idx = df[['신중도_A','신중도_B','신중도_C','max_number']]

def match_num(x):
    if x[0] == x[3]:
        return '신중도_A'
    elif x[1] == x[3]:
        return '신중도_B'
    elif x[2] == x[3]:
        return '신중도_C'
```

## 신중도 끝

```python
idx['신중도_value'] = idx.apply(match_num,axis=1)
idx
```

```python
# count A,B,C
A_list=[]
B_list=[]
C_list=[]
D_list=[]

for i in range(df2.shape[0]):
    A_list.append(list(df2.iloc[i]).count('A'))
    B_list.append(list(df2.iloc[i]).count('B'))
    C_list.append(list(df2.iloc[i]).count('C'))
    D_list.append(list(df2.iloc[i]).count('D'))

df2['습관도_A'] = A_list
df2['습관도_B'] = B_list
df2['습관도_C'] = C_list
df2['습관도_D'] = D_list
```

```python
idx2 = df2[['습관도_A','습관도_B','습관도_C','습관도_D','max_number']]

def match_num(x):
    if x[0] == x[4]:
        return '습관도_A'
    elif x[1] == x[4]:
        return '습관도_B'
    elif x[2] == x[4]:
        return '습관도_C'
    elif x[3] == x[4]:
        return '습관도_D'
   
```
