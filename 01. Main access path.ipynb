```python
# import 
import numpy as np
import pandas as pd
import os

from datetime import datetime
```

```python
action_df = pd.read_csv('./data/제6회 L.POINT Big Data Competition-분석용데이터-01.온라인 행동 정보.csv')
```

```python
# datetime으로 변경
action_df.sess_dt = action_df.sess_dt.astype(str)
action_df.hit_tm = action_df.hit_tm.astype(str)
action_df['date'] = action_df.sess_dt +' '+action_df.hit_tm
action_df['date'] = action_df['date'].map(lambda x: datetime.strptime(x,'%Y%m%d %H:%M'))
# to Datetime index 
action_df.sess_dt = pd.to_datetime(action_df.sess_dt)
action_df = action_df.sort_values('date',ascending=True)
action_df = action_df.set_index(['date'])
```

```python
idx = action_df[['clnt_id','biz_unit','trfc_src','dvc_ctg_nm']]
idx.head()
```

# 주된매체 구하기 

```python
# drop_duplicates
idx = idx.reset_index()
idx = idx.drop_duplicates(['clnt_id','biz_unit','trfc_src','dvc_ctg_nm'])

not_list = []


for i in list(idx.clnt_id.unique()):
    aa = idx[idx.clnt_id == i]
    if len(aa) != 1:
        not_list.append(i)
    else: 
        pass
    
in_list = []

for i in list(idx.clnt_id.unique()):
    if i not in not_list:
        in_list.append(i)
```

```python
# 복합적으로 가지고 있는 사람들 

trfc_src_count = []
dvc_ctg_nm_count = []
biz_unit_count =[]

trfc_src_main = []
dvc_ctg_nm_main = []
biz_unit_main =[]


for i in not_list:
    aa = idx[idx.clnt_id == i]
    trfc_src_count.append(len(aa.trfc_src.unique()))
    dvc_ctg_nm_count.append(len(aa.dvc_ctg_nm.unique()))
    biz_unit_count.append(len(aa.biz_unit.unique()))
    
    
    
    trfc_src_main.append(list(aa.trfc_src.unique()))
    dvc_ctg_nm_main.append(list(aa.dvc_ctg_nm.unique()))
    biz_unit_main.append(list(aa.biz_unit.unique()))
```

```python
complicated_person  = pd.DataFrame()

complicated_person['clnt_id'] = not_list
complicated_person['trfc_src_count'] = trfc_src_count
complicated_person['dvc_ctg_nm_count'] = dvc_ctg_nm_count
complicated_person['biz_unit_count'] = biz_unit_count

complicated_person['trfc_src_main'] = trfc_src_main
complicated_person['dvc_ctg_nm_main'] = dvc_ctg_nm_main
complicated_person['biz_unit_cmain'] = biz_unit_main
```

```python
complicated_person
```

```python
non_biz_unit_count =[]
non_trfc_src_main = []
non_dvc_ctg_nm_main =[]
non_biz_unit_main =[]

for i in in_list:
    aa = idx[idx.clnt_id == i]

    non_biz_unit_count.append(len(aa.biz_unit.unique()))
    
    non_trfc_src_main.append(list(aa.trfc_src.unique()))
    non_dvc_ctg_nm_main.append(list(aa.dvc_ctg_nm.unique()))
    non_biz_unit_main.append(list(aa.biz_unit.unique()))

```

```python
no_complicated_person = pd.DataFrame()

no_complicated_person['clnt_id'] = in_list
no_complicated_person['trfc_src_count'] = 1
no_complicated_person['dvc_ctg_nm_count'] = 1
no_complicated_person['biz_unit_count'] = non_biz_unit_count


no_complicated_person['trfc_src_main'] = non_trfc_src_main
no_complicated_person['dvc_ctg_nm_main'] = non_dvc_ctg_nm_main
no_complicated_person['biz_unit_cmain'] = non_biz_unit_main

```

```python
no_complicated_person.head()
```

```python
meache_df = pd.concat([no_complicated_person,complicated_person],axis=0)
meache_df
```
